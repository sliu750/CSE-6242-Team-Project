**Combine and clean data for all air quality csv files**

```python
air_quality_df = pd.read_csv('Air_Quality_Data_2012_2024.csv')
```

```python
air_quality_df.head()
```

```python
urban_air_quality_and_health_df = pd.read_csv('Urban Air Quality and Health Impact Dataset.csv')
```

```python
urban_air_quality_and_health_df.head()
```

```python
air_quality_combined_df = pd.merge(air_quality_df, urban_air_quality_and_health_df, on="City", how="inner")
```

```python
air_quality_combined_df = air_quality_combined_df.set_index("City").reset_index()
```

```python
pd.set_option('display.max_columns', None)
air_quality_combined_df.head()
```

Remove unnecessary columns from air_quality_combined_df

```python
air_quality_combined_df = air_quality_combined_df.drop(columns=['Date', 'datetime', 'datetimeEpoch', 'dew', 'precip', 'precipprob', 'precipcover', 'preciptype', 'snow', 'snowdepth', 'windgust', 'winddir',
                                                                'pressure', 'cloudcover', 'solarradiation', 'solarenergy', 'sunrise', 'sunriseEpoch', 'sunset', 'sunsetEpoch', 'moonphase', 'stations', 'source',
                                                                'Temp_Range', 'Condition_Code', 'Month', 'Season', 'Day_of_Week', 'Is_Weekend'])
```

```python
air_quality_combined_df.head()
```

```python
%cd /content/drive/MyDrive/DVA 6242 Team Project/Datasets/
air_quality_combined_df.to_csv('AIR_QUALITY_COMBINED.csv', index=False)
```
