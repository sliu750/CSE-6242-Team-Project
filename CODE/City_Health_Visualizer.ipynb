```python
import pandas as pd
```

**Load data from datasets.**

```python
urban_air_quality_and_health_df = pd.read_csv("Urban Air Quality and Health Impact Dataset.csv")
```

```python
urban_air_quality_and_health_df
```

```python
# columns that we are interested in visualizing
features = ["datetime", "City", "Health_Risk_Score"]
```

```python
urban_air_quality_and_health_df = urban_air_quality_and_health_df[features]
```

```python
urban_air_quality_and_health_df
```

**Compute average health risk score by city and date, which we will then visualize.**

```python
# Calculate the average health risk score by day and city
avg_health_risk_df = urban_air_quality_and_health_df.groupby(['datetime', 'City'])['Health_Risk_Score'].mean().reset_index()
avg_health_risk_df.rename(columns={'Health_Risk_Score': 'Avg_Health_Risk_Score'}, inplace=True)
```
