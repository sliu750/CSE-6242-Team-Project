```python
import numpy as np
import pandas as pd
from tqdm.notebook import tqdm
```

```python
df = pd.read_csv("ALL_TREES_DATASET.csv", low_memory=False)
df["id"] = [i for i in range(len(df.index))]

cities = set(df["city"])
for city in tqdm(cities):
    oldDf = df[df["city"] != city]
    newDf = df[df["city"] == city]
    centerLong, centerLat = newDf["longitude_coordinate"].mean(), newDf["latitude_coordinate"].mean()
    newDf["distance"] = ((newDf["longitude_coordinate"] - centerLong) ** 2 + (newDf["latitude_coordinate"] - centerLat) ** 2) ** (1 / 2)
    medianDistance = newDf["distance"].median()
    newDf = newDf[newDf["distance"] <= (medianDistance * 20)]
    newDf.drop("distance", axis=1, inplace=True)

    df = pd.concat([oldDf, newDf], axis=0)
```

```python
df.to_csv("FILTERED_ALL_TREES_DATASET.csv", index=False)
```
